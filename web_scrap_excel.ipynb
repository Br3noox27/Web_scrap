```python
import requests
from bs4 import BeautifulSoup

headers = {'user-agent':'Mozilla/5.0'}

oq_vc_quer = input('O que você quer procurar?')
numero_de_paginas = 1

lista_de_produtos = []
while numero_de_paginas > 0:
    url = f'https://sp.olx.com.br/?o={numero_de_paginas}&q={oq_vc_quer}' 
    print(' ')

    url_desejada = requests.get(url, headers=headers)
    site = BeautifulSoup(url_desejada.text, 'html.parser')

    Produtos = site.findAll('li', attrs={'class':'sc-1fcmfeb-2 dvcyfD'})

    for produtos in Produtos:
        
        try:
            qnts_paginas = site.find('span', attrs={'sc-1fcmfeb-2 dvcyfD'}).text.replace('1 - 50 de ', '')
        except:
            qnts_paginas = ''
            
        print(qnts_paginas)
        
        try:

            nome = produtos.find('h2', attrs={'kgl1mq-0 eFXRHn sc-ifAKCX iUMNkO'})
            lugar = produtos.find('span', attrs={'sc-1c3ysll-1 cLQXSQ sc-ifAKCX lgjPoE'})
            preco = produtos.find('span', attrs={'m7nrfa-0 eJCbzj sc-ifAKCX jViSDP'})
            try:
                link = produtos.find('a', attrs={'sc-12rk7z2-1 huFwya sc-gzVnrw gEAEAL'})['href']
            except:
                link = ''

            print('link:', link)
            print("Titulo:", nome.text)
            print('Preço do Produto:', preco.text)
            print('Lugar:', lugar.text)

            dicionario = {
                'link': link,
                'titulo': nome.text,
                'preço do produto': preco.text,
                'lugar': lugar.text
            }

            lista_de_produtos.append(dicionario)

            print('')

        except AttributeError as erro:
            print('erro', erro)

    numero_de_paginas +=1
```

```python
import pandas as pd

# Crie um DataFrame a partir da lista de produtos
df = pd.DataFrame(lista_de_produtos)

# Escreva o DataFrame em um arquivo Excel
df.to_excel('moletom_adidas.xlsx', index=False)
```
